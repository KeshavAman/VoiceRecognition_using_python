```python
import pyttsx3
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser
import os
from googlesearch import search
```

```python
engine = pyttsx3.init()
voices = engine.getProperty('voices')
#print(voices[1].id)
engine.setProperty('voice',voices[0].id)
```

```python
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
```

```python
def wishMe():
    hour = int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("Good Morning !")
        
    elif hour>=12 and hour<18:
        speak("Good Afternoon !")
        
    else:
        speak("Good Evening !")
    speak("I am Jarvis. How may I help you")
```

```python
def takeCommand():
    #It takes microphone input from the user and returns string output

    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source)
    try:
        print("Recognizing...")    
        query = r.recognize_google(audio, language='en-in') #Using google for voice recognition.
        print(f"User said: {query}\n")  #User query will be printed.

    except Exception as e:
        # print(e)    
        print("Say that again please...")  #Say that again will be printed in case of improper voice 
        speak('say that again please...')
        return "None" #None string will be returned
    return query
 
```

```python
if __name__=="__main__":
    wishMe()
    while True:
        query = takeCommand().lower()
        
        if 'wikipedia' in query:
            speak('Searching Wikipedia...')
            query = query.replace("wikipedia", "")
            results = wikipedia.summary(query, sentences = 3)
            speak("According to Wikipedia")
            print(results)
            speak(results)
            
        elif 'open youtube' in query:
            webbrowser.open('https://www.youtube.com')
            break
        
        elif 'open google' in query:
            webbrowser.open('https://www.google.com')
            break
            
        elif 'stop' in query or 'quit' in query or 'exit' in query or 'bye' in query or 'tata' in query:
            speak('Thanks for talking to me see you later ')
            break
            
        elif 'the time' in query:
            strTime = datetime.datetime.now().strftime("%H:%M:%S")
            print(strTime)
            speak(f"The time is {strTime}")
            
        elif 'open photos' in query:
            path = "/Applications/Photos.app"
            os.system(f"open {path}")
            break
            
        elif 'open safari' in query:
            path = "/Applications/Safari.app"
            os.system(f"open {path}")
            break
            
        elif 'open music' in query:
            path = "/Applications/iTunes.app"
            os.system(f"open {path}")
            break
            
        elif 'open facetime' in query:
            path = "/Applications/FaceTime.app"
            os.system(f"open {path}")
            break
         
        elif 'open chrome' in query:
            path = "/Applications/Google Chrome.app"
            os.system(f"open {path}")
            break
            
        elif "will you be my gf" in query or "will you be my bf" in query or "will you be my boyfriend" in query or "will you be my girlfriend" in query:  
            speak("I'm not sure about, may be you should give me some time")
 
        elif "i love you" in query:
            speak("It's hard to understand")
            
        elif 'how are you' in query:
            speak("I am fine, Thank you")
            speak("How are you")
 
        elif 'fine' in query or "good" in query:
            speak("It's good to know that your fine")
 
        elif "what's your name" in query or "What is your name" in query:
            print("My friends call me jarvis")
            speak("My friends call me jarvis")
            
        elif 'hey jarvis' in query or 'hye jarvis' in query or 'hello jarvis' in query or 'hello' in query:
            print('Hi Go ahead')
            speak('Hi Go ahead')
            
        elif 'google' in query:
            for j in search(query, tld="co.in", num=10, stop=10, pause=2):
                print(j)
            break
            
        elif 'tell me about yourself' in query:
            speak("Hello my name is jarvis and i am your virtual assistant and also i am here to help you")
```
